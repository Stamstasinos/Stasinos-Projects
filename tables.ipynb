```python
filtered_df = df[(df["vehicles_above_q90"] == 1) & (df["year"] == 2019)]

# Drop rows with missing values in any of the key columns
filtered_df = filtered_df.dropna(subset=["region", "eurosceptic_vote_share", "euroscepticism_change"])

# Select and rename columns for readability
filtered_df = filtered_df[["region", "eurosceptic_vote_share", "euroscepticism_change"]]
filtered_df.columns = ["Region", "Eurosceptic Vote Share", "Euroscepticism Change in %"]

# Round to two decimals
filtered_df["Eurosceptic Vote Share"] = filtered_df["Eurosceptic Vote Share"].round(2)
filtered_df["Euroscepticism Change in %"] = filtered_df["Euroscepticism Change in %"].round(2)

# Sort descending by Eurosceptic Vote Share and reset index
filtered_df = filtered_df.sort_values(by="Eurosceptic Vote Share", ascending=False).reset_index(drop=True)
filtered_df.index += 1

# Print the table
print(tabulate(filtered_df, headers='keys', tablefmt='pretty', showindex=True))
```

```python
regions_data_agri = df[(df['agri_above_q90'] == 1) & (df['year'] == 2019)][
    ['region', 'eurosceptic_vote_share', 'euroscepticism_change']
].drop_duplicates().reset_index(drop=True)

# Ensure numeric conversion
regions_data_agri['eurosceptic_vote_share'] = pd.to_numeric(regions_data_agri['eurosceptic_vote_share'], errors='coerce')
regions_data_agri['euroscepticism_change'] = pd.to_numeric(regions_data_agri['euroscepticism_change'], errors='coerce')

# Round to 2 decimals
regions_data_agri['eurosceptic_vote_share'] = regions_data_agri['eurosceptic_vote_share'].round(2)
regions_data_agri['euroscepticism_change'] = regions_data_agri['euroscepticism_change'].round(2)

# Drop NaNs
regions_data_agri = regions_data_agri.dropna()

# Sort by vote share descending
regions_data_sorted_agri = regions_data_agri.sort_values(by='eurosceptic_vote_share', ascending=False).reset_index(drop=True)
regions_data_sorted_agri.index += 1  # Optional: start index from 1

# Print as table
print(tabulate(regions_data_sorted_agri.rename(columns={
    'group_id': 'NUTS 2',
    'region': 'Region',
    'eurosceptic_vote_share': 'Eurosceptic Vote Share',
    'euroscepticism_change': 'Euroscepticism Change in %'
}), headers='keys', tablefmt='pretty', showindex=True))
```

```python
filtered_df = df[df['year'] == 2019][
    ['group_id','region', 'gdp_per_capita', 'eurosceptic_vote_share', 'euroscepticism_change', 
     'agri_above_eu_mean', 'vehicles_above_eu_mean']
].dropna()

# Sort by GDP per capita (ascending) and select top 20
lowest_gdp_regions = filtered_df.sort_values(by='gdp_per_capita').head(20).reset_index(drop=True)

# Round numeric values
lowest_gdp_regions['gdp_per_capita'] = lowest_gdp_regions['gdp_per_capita'].round(2)
lowest_gdp_regions['eurosceptic_vote_share'] = lowest_gdp_regions['eurosceptic_vote_share'].round(2)
lowest_gdp_regions['euroscepticism_change'] = lowest_gdp_regions['euroscepticism_change'].round(2)

# Rename columns for presentation
lowest_gdp_regions = lowest_gdp_regions.rename(columns={
    'region': 'Region',
    'gdp_per_capita': 'GDP per Capita',
    'eurosceptic_vote_share': 'Eurosceptic Vote Share',
    'euroscepticism_change': 'Euroscepticism Change',
    'agri_above_eu_mean': 'Agri > EU Mean',
    'vehicles_above_eu_mean': 'Vehicles > EU Mean'
})

# Add a column counting how many times the region is above EU mean in agri and vehicle indicators
lowest_gdp_regions['Above EU Mean Count'] = (
    lowest_gdp_regions[['Agri > EU Mean', 'Vehicles > EU Mean']].sum(axis=1)
)

# Print as table
print(tabulate(lowest_gdp_regions, headers='keys', tablefmt='pretty', showindex=True))
```
