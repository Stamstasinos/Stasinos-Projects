```python
filtered_df = df[df['year'] == 2019][
    ['group_id','region', 'gdp_per_capita', 'eurosceptic_vote_share', 'euroscepticism_change', 
     'agri_above_eu_mean', 'vehicles_above_eu_mean']
].dropna()

# Sort by GDP per capita (ascending) and select top 20
lowest_gdp_regions = filtered_df.sort_values(by='gdp_per_capita').head(20).reset_index(drop=True)

# Round numeric values
lowest_gdp_regions['gdp_per_capita'] = lowest_gdp_regions['gdp_per_capita'].round(2)
lowest_gdp_regions['eurosceptic_vote_share'] = lowest_gdp_regions['eurosceptic_vote_share'].round(2)
lowest_gdp_regions['euroscepticism_change'] = lowest_gdp_regions['euroscepticism_change'].round(2)

# Rename columns for presentation
lowest_gdp_regions = lowest_gdp_regions.rename(columns={
    'region': 'Region',
    'gdp_per_capita': 'GDP per Capita',
    'eurosceptic_vote_share': 'Eurosceptic Vote Share',
    'euroscepticism_change': 'Euroscepticism Change',
    'agri_above_eu_mean': 'Agri > EU Mean',
    'vehicles_above_eu_mean': 'Vehicles > EU Mean'
})

# Add a column counting how many times the region is above EU mean in agri and vehicle indicators
lowest_gdp_regions['Above EU Mean Count'] = (
    lowest_gdp_regions[['Agri > EU Mean', 'Vehicles > EU Mean']].sum(axis=1)
)

# Print as table
print(tabulate(lowest_gdp_regions, headers='keys', tablefmt='pretty', showindex=True))
```

```python
filtered_df = df[df['year'] == 2019][
    ['group_id','region', 'gdp_per_capita', 'eurosceptic_vote_share', 'euroscepticism_change', 
     'agri_above_eu_mean', 'vehicles_above_eu_mean']
].dropna()

# Sort by GDP per capita (ascending) and select top 20
lowest_gdp_regions = filtered_df.sort_values(by='gdp_per_capita').head(50).reset_index(drop=True)

# Round numeric values
lowest_gdp_regions['gdp_per_capita'] = lowest_gdp_regions['gdp_per_capita'].round(2)
lowest_gdp_regions['eurosceptic_vote_share'] = lowest_gdp_regions['eurosceptic_vote_share'].round(2)
lowest_gdp_regions['euroscepticism_change'] = lowest_gdp_regions['euroscepticism_change'].round(2)

# Rename columns for presentation
lowest_gdp_regions = lowest_gdp_regions.rename(columns={
    'region': 'Region',
    'gdp_per_capita': 'GDP per Capita',
    'eurosceptic_vote_share': 'Eurosceptic Vote Share',
    'euroscepticism_change': 'Euroscepticism Change',
    'agri_above_eu_mean': 'Agri > EU Mean',
    'vehicles_above_eu_mean': 'Vehicles > EU Mean'
})

# Add a column counting how many times the region is above EU mean in agri and vehicle indicators
lowest_gdp_regions['Above EU Mean Count'] = (
    lowest_gdp_regions[['Agri > EU Mean', 'Vehicles > EU Mean']].sum(axis=1)
)

# Print table
print(tabulate(lowest_gdp_regions, headers='keys', tablefmt='pretty', showindex=True))

# Print total number of '1's in each indicator column
agri_total = lowest_gdp_regions['Agri > EU Mean'].sum()
vehicles_total = lowest_gdp_regions['Vehicles > EU Mean'].sum()
both_total = lowest_gdp_regions['Above EU Mean Count'].sum()

print(f"\nTotal with Agri > EU Mean = 1: {agri_total}")
print(f"Total with Vehicles > EU Mean = 1: {vehicles_total}")
print(f"Total 'Above EU Mean Count' (sum of both per region): {both_total}")
```
